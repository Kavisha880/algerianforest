```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_set=pd.read_csv('Algerian_forest_fires_dataset.csv',header=1)
```

```python
data_set.head()
```

```python
data_set.info()
```

```python
# data cleaning
data_set[data_set.isnull().any(axis=1)]
```

```python
data_set.loc[:122,"Region"]=0
data_set.loc[122:,"Region"]=1
df=data_set
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
data_set[data_set.isnull().any(axis=1)]
```

```python
# remove nulll val
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# 122 is repeated row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.iloc[[121]]
```

```python
df.columns
```

```python
# fix spaces in columns
df.columns=df.columns.str.strip()
df.columns
```

```python
# change the required columns as integer datatype

```

```python
df[['month', 'year', 'Temperature', 'RH', 'Ws']]=df[['month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)

# objects=[features for features in df.columns if df[features].dtypes=='O' ]
# objects
```

```python
# data Analysis

df_copy=df.drop(['day','month', 'year'],axis=1)
# what to predict :: FWI
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
# visualisation
# plot density for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# % of pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# pie chart
import matplotlib.pyplot as plt

classLabel=["Fire","NotFire"]
plt.figure(figsize=(12, 7))
plt.pie(percentage,labels=classLabel,autopct='%1.1f%%')
plt.title("Pie chart of classes")
plt.show()
```

```python
# correlation
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
plt.show()
```

```python
# fwi :dependent feature
sns.boxplot(df['FWI'])
plt.show()
```

```python
# monthly fire analysis
dftemp =df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel("No of fires", weight="bold")
plt.xlabel("Months", weight="bold")
plt.title("Sidi-Bel Abbes Region")

plt.show()
```

```python
# monthly fire analysis
dftemp =df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel("No of fires", weight="bold")
plt.xlabel("Months", weight="bold")
plt.title("bajaja region data_set")

plt.show()
```
